```python
import networkx as nx
import seaborn as sns
import numpy as np
import scipy as sp
import pandas as pd
from collections import defaultdict
```

```python
df = pd.read_csv('data/valuesMatrix.csv')
df.columns
df.head(10)
```

```python
nameDict = df['name']
nameDict[0]
df = df.drop('name',axis=1)
# # range of lvls 1-5
df[df['lvl'] < 5] -= 1
# newCol = df['lvl'].to_numpy()
# newCol = [[x] for x in newCol]
# df['lvl'] = newCol
df.head(10)
```

```python

def createAdjList(df):
    adjDictScale = defaultdict(list)
    adjDictValue = defaultdict(list)
    #DICT
    for idx, name in enumerate(df):
        # connect scale nodes to create complete graph
        if idx < 4:
            adjDictValue[idx] = (idx+1, idx+1, name)
       # fully connected graph
        elif idx == 5:
             adjDictValue[idx] = (idx+1, 0, name) 
        else:
            adjDictValue[idx] = (idx+1,df.iloc[idx][0], name)
            
    return (adjDictValue,adjDictScale)


def createScaleAndValue(df):
    
    adjDictScale = defaultdict(list)
    adjDictValue = defaultdict(list)
    #DICT
    for idx, name in enumerate(df):
        # connect scale nodes to create complete graph
        if idx < 4:
            adjDictScale[idx] = (idx+1, idx+1, name)
       # fully connected graph
        elif idx == 5:
             adjDictScale[idx] = (idx+1, 0, name) 
        else:
            adjDictValue[idx] = (idx+1,df.iloc[idx][0], name)
            
    return (adjDictValue,adjDictScale)

def bipartite(u,v,edges):
    bi = nx.Graph()
    bi.add_nodes_from(u, bipartite=0)
    bi.add_nodes_from(v, bipartite=1)
    bi.add_edges_from(edges)
    return bi



```

```python
# Call G1
g1_adjDictValue, g1_adjDictScale = createAdjList(df)
g2_adjDictValue, g2_adjDictScale = createScaleAndValue(df)

g1_adjDictValue
```
