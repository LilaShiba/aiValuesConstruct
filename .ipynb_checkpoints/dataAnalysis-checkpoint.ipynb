```python

def createAdjList(df):
    adjDictScale = defaultdict(list)
    adjDictValue = defaultdict(list)
    #DICT
    for idx, name in enumerate(df):
        # connect scale nodes to create complete graph
        if idx < 4:
            adjDictValue[idx] = (idx+1, idx+1, name)
       # fully connected graph
        elif idx == 5:
             adjDictValue[idx] = (idx+1, 0, name) 
        else:
            adjDictValue[idx] = (idx+1,df.iloc[idx][0], name)
            
    return (adjDictValue,adjDictScale)


def createScaleAndValue(df):
    
    adjDictScale = defaultdict(list)
    adjDictValue = defaultdict(list)
    #DICT
    for idx, name in enumerate(df):
        # connect scale nodes to create complete graph
        if idx < 4:
            adjDictScale[idx] = (idx+1, idx+1, name)
       # fully connected graph
        elif idx == 5:
             adjDictScale[idx] = (idx+1, 0, name) 
        else:
            adjDictValue[idx] = (idx+1,df.iloc[idx][0], name)
            
    return (adjDictValue,adjDictScale)

def bipartite(u,v,edges):
    bi = nx.Graph()
    bi.add_nodes_from(u, bipartite=0)
    bi.add_nodes_from(v, bipartite=1)
    bi.add_edges_from(edges)
    return bi


# Call
g1_adjDictValue, g1_adjDictScale = createAdjList(nameDict)
g2_adjDictValue, g2_adjDictScale = createScaleAndValue(nameDict)

# scale
u = [x for x in range(4)]
v = [x for x in range(4,104)]
edges = [(edge,scale) for edge,scale in enumerate(g2_adjDictValue)]

biGraph = bipartite(u,v,edges)


# Array

# adjListScale = list()
# adjListValue = list()

# for idx, name in enumerate(nameDict):
#     if idx < 5:
#         adjListScale.append( (idx ,df.iloc[idx][0]) )
#     else:
#         adjListValue.append((idx, df.iloc[idx][0]))
        
```
